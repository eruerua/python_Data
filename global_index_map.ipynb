```python
import tushare as ts
from pyecharts import options as opts
from pyecharts.charts import Map
```

```python
index_key = {
    'SPTSX': 'Canada',
    'FCHI': 'France',
    'KS11': 'South Korea',
    'SPX': 'United States',
    'GDAXI': 'Germany',
    'N225':'Japan',
    'AS51':'Australia',
    'SENSEX': 'India',
    'IBOVESPA':'Brazil',
    'RTS':'Russia',
    'TWII':'Taiwan',
    'HSI':'Hong Kong'
}
```

```python
pro = ts.pro_api('')
#获取全球指数
df = pro.index_globaL(trade_date='20200324', fields='pct_chg,ts_code')
df['country'] = df['ts_code'].map(index_key)
df = df.dropna(subset=['country'])
df = df[['ts_code', 'pct_chg', 'country']]
```

```python
#获取上证指数，并合并
sh = pro.index_daily(ts_code='000001.SH', fields='ts_code,pct_chg,ts_code', trade_date='20200220')
sh['country'] = 'China'
data = df.append(sh)
```

```python
data
```

```python
def index_global():
    c = (
        Map()
        .add("", 
             data_pair=[list(z) for z in zip(data['country'].values, data['pct_chg'].values)], 
             maptype="world",
             is_map_symbol_show=False)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="国际主要指数表现"),
            visualmap_opts=opts.VisualMapOpts(max_=2,
                                              min_=-2,
                                              is_calculable=True,
                                             range_color = ['green', 'red']),
        )
    )
    return c
index_global().render_notebook()
```
